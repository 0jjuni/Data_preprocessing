```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
```

### 데이터 준비

- sklearn에 내장된 데이터인 iris 데이터를 이용

```python
iris = load_iris()

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
```

## 1.사이킷런의 OneHotEncoder 이용

```python
#2차원 행렬로 변환
encoding_target = iris_df['species'].values.reshape(-1,1)

# OneHotEncoder 객체 생성 및 적용
encoder = OneHotEncoder()
onehot_encoded1 = encoder.fit_transform(encoding_target)

#인코딩 결과 확인
onehot_encoded1 = onehot_encoded1.toarray()
print(onehot_encoded1)
```

```python
# 기존 데이터프레임과 원핫 인코딩 결과를 합치기
temp = pd.DataFrame(onehot_encoded1, columns=iris.target_names)
iris_df_encoded1 = pd.concat([iris_df, temp], axis=1)

#결과 확인
display(iris_df_encoded1)
```

## 2. Pandas의 get_dummies 이용

```python
# 'species' 열을 원핫 인코딩
onehot_encoded2 = pd.get_dummies(iris_df['species'])

#인코딩 결과 확인
print(onehot_encoded2)
```

```python
# 기존 데이터프레임과 원핫 인코딩 결과를 합치기
iris_df_encoded2 = pd.concat([iris_df, onehot_encoded2], axis=1)

# 결과 확인
display(iris_df_encoded2)
```
